```python
hsv = cv2.cvtColor(img_minha, cv2.COLOR_RGB2HSV) ### hue (matiz, tonalidade) - saturação - valor (brilho)
plt.imshow(hsv);
```

```python
brilho = 67
hsv[:,:,2] = cv2.add(hsv[:,:,2], brilho)
img_minha = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
plt.imshow(img_minha);
```

```python
brilho = 50
hsv[:,:,1] = cv2.add(hsv[:,:,1], brilho)
img_minha = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
plt.imshow(img_minha);
```

```python
height, width = img_minha.shape[:2]
center = (width/2, height/3)
rotate_matrix = cv2.getRotationMatrix2D(center=center, angle=45, scale=2)
rotated_image = cv2.warpAffine(src=img_minha, M=rotate_matrix, dsize=(width, height))
plt.imshow(rotated_image);
```

```python
M = np.float32([[0.5, 0.5, 0],
                [0, 0.5 , 0],
                [0, 0  , 1]])
M = np.float32([[1,   0, 0],
                [0.1, 0.5, 0],
                [0,   0, 1]])
rows, cols, dim = img_minha.shape
sheared_img = cv2.warpPerspective(img_minha,M,dsize=(width, height))
plt.imshow(sheared_img);
```
