## Exercício de Concatenar, mergear e afins no |DataFrame|


```python
novo_passageiro = {
    "PassengerId": [892],
    "Survived": [1],
    "Pclass": [3],
    "Name": ['Souza Machado, Rafael R.'],
    "Sex": ['male'],
    "Age": [28],
    "SibSp": [0],
    "Parch": [0],
    "Ticket": [56784],
    "Fare": [7.75],
    "Cabin": ['B12'],
    "Embarked": ['Q'],
}
```

```python
df_novo_passageiro = pd.DataFrame(novo_passageiro)
df_novo_passageiro
```

```python
pd.concat([base_dados, df_novo_passageiro], ignore_index=True)
```

```python
novas_colunas = {
    'coluna1': [i for i in range(base_dados.shape[0])],
    'coluna2': [i + 50 for i in range(base_dados.shape[0])]

}

df_novas_colunas = pd.DataFrame(novas_colunas)
df_novas_colunas.head()
```

```python
pd.concat([base_dados, df_novas_colunas], axis=1)
```

```python
base_dados_alt = pd.read_csv('titanic_altura.csv')
base_dados_alt.head()
```

```python
base_dados_alt = base_dados_alt.sample(base_dados_alt.shape[0])
base_dados_alt.head()
```

```python
pd.merge(left=base_dados, right=base_dados_alt, on='PassengerId', how='inner')
```
