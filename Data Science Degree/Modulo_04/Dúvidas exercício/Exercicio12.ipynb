```python
import numpy as np
from cmath import exp, sin
import matplotlib.pyplot as plt

x = np.random.randint(-10, 10, size=200)

y1 = []
y2 = []
for i in range(len(x)):
    x_n = x[i]
    if x_n == 0:
        y1.append(0)
        y2.append(0)
    else:
        y1.append(sin(x_n)/x_n)

        y2.append(1/(1-exp(-x_n)))

plt.plot(x, y1)
plt.show()

plt.plot(x, y2)
plt.show()
```
